```python
import matplotlib.pyplot as plt
df=df_1


plt.figure(figsize=(10,5))
plt.plot(df['Age'],df['Income'],color='red',marker='o')
plt.title('Age vs Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
df=df.sort_values(by='Age')
plt.figure(figsize=(10,5))
plt.plot(df['Age'],df['Income'],color='red',marker='*',markersize=10)
plt.title('Age vs Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
df=df.sort_values(by='Name')
plt.figure(figsize=(10,5))
plt.plot(df['Name'],df['Income'],color='red',marker='*',markersize=10)
plt.title('Name vs Income')
plt.xlabel('Name')
plt.ylabel('Income')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(df['City'],df['Income'],color='red')
plt.title('Amount by Category')
plt.xlabel('City')
plt.ylabel('Amount')
plt.show()
```
